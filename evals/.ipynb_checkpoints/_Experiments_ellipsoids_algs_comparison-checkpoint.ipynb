```python
import sys
import numpy as np
import cvxpy as cp
import mosek

def Solve_Mosek(x_p, A2, b2, c2, A3, b3, c3):
    size = np.size(x_p)
    x = cp.Variable((size))

    constraints = [
        cp.quad_form(x, A2) + b2 @ x + c2 <= 0,
        cp.quad_form(x, A3) + b3 @ x + c3 <= 0
    ]

    objective = cp.Minimize(cp.norm(x - x_p) ** 2)
    prob = cp.Problem(objective, constraints)
    tol = 1e-4
    prob.solve(solver=cp.MOSEK, mosek_params = {mosek.dparam.intpnt_co_tol_rel_gap: tol,
                              mosek.dparam.intpnt_co_tol_dfeas: tol,
                              mosek.dparam.intpnt_co_tol_infeas: tol * 1.0e-2,
                              mosek.dparam.intpnt_co_tol_mu_red: tol,
                              mosek.dparam.intpnt_co_tol_pfeas: tol,
                              }
              )

    return prob
```

```python
epsilon = 1e-4
N = 5
k = 9

times_ell = np.zeros((k,N))
times_ell_2 = np.zeros((k,N))
times_ell_3 = np.zeros((k,N))
times_slsqp = np.zeros((k,N))
times_ipm = np.zeros((k,N))
times_admm = np.zeros((k,N))
times_FW = np.zeros((k,N))


for i in range(N):
    j = 0
    for n in (10,100,500,1000,2000,5000,8000,10000,11000):
        print(" ")
        print("n = ", n)

        A = np.random.random((n,n)) / n
        A_norm = np.linalg.norm(A + np.eye(n) / n)
        A2_half = (A + np.eye(n) / n) / A_norm**0.5
        A2 = A2_half.T.dot(A2_half)
        x_c2 = np.random.random(n) / n
        b2 = -2. * A2_half.T.dot(x_c2)
        c2 = x_c2.dot(x_c2) - 1.

        
        A = np.random.random((n,n)) / n
        A_norm = np.linalg.norm(A + np.eye(n) / n)
        A3_half = (A + np.eye(n)/n) / A_norm**0.5
        A3 = A3_half.T.dot(A3_half)
        x_c3 = np.random.random(n) / n
        b3 = -2. * A3_half.T.dot(x_c3)
        c3 = x_c3.dot(x_c3) - 1.
        
        ############

        x_0 = np.zeros(n)
        x_p = np.zeros(n)
        x_p[0] = 100.
        
######## SLSQP #############
        print("SLSQP")
        if n <= 5000:
            fun = lambda x: np.linalg.norm(x - x_p)**2 
            cons = ({'type': 'ineq', 'fun': lambda x: -x.dot( A2.dot(x)) -  b2.dot(x) -  c2},
                    {'type': 'ineq', 'fun': lambda x: -x.dot( A3.dot(x)) -  b3.dot(x) -  c3})
            t0 = time.time()
            SLSQP_sol = minimize(fun, x_0, method = 'SLSQP',  constraints=cons, tol = epsilon)
            times_slsqp[j,i] = time.time() - t0
            SLSQP_res = SLSQP_sol.x

        print("time_slsqp = ", times_slsqp[j,i])
        
######## Ellipsoid #############
        print("Ellipsoid")
        params = (0, 0, 0)
        L = 2 
        learning_rate = 1. / 2. / L
        
        def h_constraint2(x, params):
            h_con = np.zeros(2)
            h_con[0] =  x.dot(A2.dot(x)) + b2.dot(x) + c2
            h_con[1] =  x.dot(A3.dot(x)) + b3.dot(x) + c3
            return h_con

        def h_grad2(x, params):
            d = x.size
            dh = np.zeros((2,d))
            dh[0] = 2 * A2.dot(x) +  b2
            dh[1] = 2 * A3.dot(x) +  b3
            return dh

        T = 500
        l_up =  15000
        params = (0, (A2,b2,c2), (A3,b3,c3))
        t1 = time.time()
        ell_sol = Ellipsoid(l_up,  epsilon, x_0, x_p, L, T, AGD_m, 
                            h_grad2, h_constraint2, learning_rate, params,
                           f_objective2, f_grad2, K_max = 10000, mu = 2 , G = 0.1)
        times_ell[j,i] = time.time() - t1
        print("time_ellipsoid = ", times_ell[j,i])
        
######## MOSEK #############
#         print("MOSEK")
        t4 = time.time()
        ipm_sol = Solve_Mosek(x_p, A2, b2, c2, A3, b3, c3)
        times_ipm[j,i] = time.time() - t4
        ipm_res = ipm_sol.solution.primal_vars
        print("time_ipm = ", times_ipm[j,i])

        j += 1
```

# ADMM

```python
def f_1(x_s, x_p, rho, A_long, c_long, z, v):
    return np.linalg.norm(x_s - x_p)**2  + rho / 2. * np.linalg.norm(A_long.dot(x_s) - c_long - z + v)**2
    
def f_grad1(x_s, x_p, rho, A_long, c_long, z, v):
    l1 = 2 * (x_s - x_p) 
    l2 = 2 * A_long.dot(x_s) - c_long - z + v
    l3 = rho / 2. * A_long.T.dot(l2)
    return  l1 + l3
    
def AGD_m_simple(x_start, x_p, 
                 rho, A_long, c_long, z, v,
                 t_epsilon, learning_rate, params, 
                 f_objective=f_1, f_grad=f_grad1, t=1, K_max = 1000, mu = 2):
    
    x_k = x_start
    y_k = x_start
    k = 0
    size = x_k.size  
    l1 = 1. / learning_rate * np.max((rho,1))
    l2 = np.min((l1, 2))
    kappa = l1 / mu
    u = (kappa**0.5 - 1.)/(kappa**0.5 + 1.) 
    
    time1 = time.time()
    grad = f_grad(x_k, x_p, rho, A_long, c_long, z, v)
    iter_time1 = time.time() - time1
    grad_norm = np.linalg.norm(grad)

    while  grad_norm > t_epsilon:
        y_kPlus1 = x_k - learning_rate * grad / np.max((rho,1))
        x_kPlus1 = (1 + u) * y_kPlus1 - u * y_k
        
        x_k = x_kPlus1
        y_k = y_kPlus1
        
        grad = f_grad(x_k, x_p, rho, A_long, c_long, z, v)
        grad_norm = np.linalg.norm(grad)

        k += 1
        
        if k > K_max:
            print("long")
            break
    return x_k, k, 1.

def proj_onto_double_unit_ball(y):
    size2 = np.size(y)
    size = int(size2 / 2.)
    y1 = y[:size]
    y2 = y[size:]
    proj1 = y1 / np.max((np.linalg.norm(y1), 1))
    proj2 = y2 / np.max((np.linalg.norm(y2), 1))
    return np.hstack((proj1, proj2))


def Solve_ADMM(x_p, x_c2, x_c3, A2_half, A3_half, 
               rho, x_0, x_opt, learning_rate, epsilon):

    t_epsilon = epsilon**2
    size = np.size(x_p)
    A_long = np.vstack(( A2_half,  A3_half))
    c_long = np.hstack(( x_c2, x_c3))
    k = 0
    x = cp.Variable((size))
    z = np.zeros(2 * size)
    v = np.zeros(2 * size)
    aug_lagr = cp.norm(x - x_p)**2  + rho / 2. * cp.norm( A_long @ x -  c_long - z + v)**2
    x.value = x_0
    while abs(np.linalg.norm(x.value - x_p)**2 - np.linalg.norm(x_opt - x_p)**2) > 2 * epsilon:
        x.value = AGD_m_simple(x_0, x_p, 
                             rho, A_long, c_long, z, v, 
                             t_epsilon, learning_rate, params, 
                             f_objective=f_1, f_grad=f_grad1, K_max = 4000, mu = 2, t = 1)[0]
        z = proj_onto_double_unit_ball(A_long.dot(x.value) - c_long + v)
        v += A_long.dot(x.value) - c_long - z
        k += 1
    print("ADMM_k", k)
    return x.value,  np.linalg.norm(x.value - x_p)**2
```

```python
n = 10
epsilon = 1e-4

A = np.random.random((n,n)) / n
A_norm = np.linalg.norm(A + np.eye(n) / n)
A2_half = (A + np.eye(n) / n) / A_norm**0.5
A2 = A2_half.T.dot(A2_half)
x_c2 = np.random.random(n) / n
b2 = -2. * A2_half.T.dot(x_c2)
c2 = x_c2.dot(x_c2) - 1.


A = np.random.random((n,n)) / n
A_norm = np.linalg.norm(A + np.eye(n) / n)
A3_half = (A + np.eye(n)/n) / A_norm**0.5
A3 = A3_half.T.dot(A3_half)
x_c3 = np.random.random(n) / n
b3 = -2. * A3_half.T.dot(x_c3)
c3 = x_c3.dot(x_c3) - 1.

x_0 = np.zeros(n)
x_p = np.zeros(n)
x_p[0] = 100.
rho = 1

fun = lambda x: np.linalg.norm(x - x_p)**2 
cons = ({'type': 'ineq', 'fun': lambda x: -x.dot( A2.dot(x)) -  b2.dot(x) -  c2},
        {'type': 'ineq', 'fun': lambda x: -x.dot( A3.dot(x)) -  b3.dot(x) -  c3})
SLSQP_sol = minimize(fun, x_0, method = 'SLSQP',  constraints=cons, tol = epsilon)
SLSQP_res = SLSQP_sol.x
x_opt = SLSQP_res
t0 = time.time()
admm_sol = Solve_ADMM(x_p, x_c2, x_c3, A2_half, A3_half, 
               rho, x_0, x_opt, learning_rate, epsilon)
print(time.time()-t0)
```
